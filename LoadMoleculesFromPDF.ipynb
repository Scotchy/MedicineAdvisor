```python
# Generate XML to create an object property link between inst1 and inst2 instances
def createOP(inst1, inst2, prop):
    if inst1 != inst2:
        return "<ObjectPropertyAssertion>\n"+ \
            "\t<ObjectProperty IRI=\"#"+prop+"\"/>\n"+\
            "\t\t<NamedIndividual IRI=\"#"+inst1+"\"/>\n"+\
            "\t<NamedIndividual IRI=\"#"+inst2+"\"/>\n"+\
        "</ObjectPropertyAssertion>"
    else:
        return ""
```

```python
# Create a link between an instance inst and all instances of a class cl
def linkInstToClass(inst, cl, prop):
    r = ""
    for c in cl:
        r+=createOP(inst, c, prop)+"\n"
    return r
```

```python
def formattxt(txt):
    mapf = ["é", "è", "ï", "î", "ë", "ê", "-", " + ", "+", " ", "'", ",", "/", "(", ")"]
    mapr = ["e", "e", "i", "i", "e", "e", "_", "_", "_", "_", "_", "", "_", "", ""]
    txt = txt.strip().lower()
    for f, r in zip(mapf, mapr):
        txt = txt.replace(f, r)
    return txt
```
